### Question 1:
Write a function named "data_preprocessing" that accepts the dataset in dataframe as only argument and does the following tasks and returns the updated dataframe:<br/>
0) Drop "PassengerId" and "Ticket" columns from the dataframe.<br/>
1) Check if there are any rows that have missing entries for all columns.<br/> 
2) Remove such rows from the given dataset.<br/>
3) Convert the data type of "Fare" and "Age" to int.<br/>
4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".<br/>
5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'.<br/>

```python
#Work on your solution here and after. 
#make sure to follow the given structure below
#1) Write your function here
def data_preprocessing (args):
    #body of the function
    df = args.drop(labels = ['PassengerId', 'Ticket'], axis = 1)
#age, cabin, and embarked are the only columns with NaN values. I would have used all for dropna but then it would give me an error to change type   
    df = df.dropna(subset = ['Age', 'Cabin','Embarked'])
    df = df.astype({'Age' : int, 'Fare': int})
    df = df.rename(columns = {'SibSp' : 'Siblings/Spouses Aboard', 'Parch' :'Parents/Children Aboard'})
    df[['Name']].applymap(lambda x: x.replace("Mlle", "Miss"))
    df[['Name']].applymap(lambda x: x.replace("Ms", "Miss"))
    df[['Name']].applymap(lambda x: x.replace("Mme", "Mrs"))
    return df

#2) Call your function here
new_titanic = data_preprocessing(titanic)
new_titanic.head()
```

```python
#Work on your solution here and after.
name_map = {'Mr' :1, 'Miss' : 2, 'Mrs' : 3, 'Master' : 4, 'Rare' : 5}
cabin_maps = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}

def adjust_age(dataset):
        dataset.loc[ dataset['Age'] <= 10, 'Age'] = 0
        dataset.loc[(dataset['Age'] > 10) & (dataset['Age'] <= 15), 'Age'] = 1
        dataset.loc[(dataset['Age'] > 15) & (dataset['Age'] <= 25), 'Age'] = 2
        dataset.loc[(dataset['Age'] > 25) & (dataset['Age'] <= 35), 'Age'] = 3
        dataset.loc[(dataset['Age'] > 35) & (dataset['Age'] <= 40), 'Age'] = 4
        dataset.loc[(dataset['Age'] > 40) & (dataset['Age'] <= 60), 'Age'] = 5
        dataset.loc[ dataset['Age'] > 60, 'Age'] = 6
        dataset['Age'] = dataset['Age'].astype(int)

            
def adjust_fare(dataset):
        dataset.loc[ dataset['Fare'] < 10, 'Fare'] = 0
        dataset.loc[(dataset['Fare'] >= 10) & (dataset['Fare'] < 205), 'Fare'] = 1
        dataset.loc[(dataset['Fare'] >= 20) & (dataset['Fare'] < 30), 'Fare'] = 2
        dataset.loc[(dataset['Fare'] >= 30) & (dataset['Fare'] < 100), 'Fare'] = 3
        dataset.loc[(dataset['Fare'] >= 100) & (dataset['Fare'] < 200), 'Fare'] = 4
        dataset.loc[ dataset['Fare'] >= 200, 'Fare'] = 5
        dataset['Fare'] = dataset['Fare'].astype(int)
            
#1) Write your function here
def feature_extraction (args):
    #body of the function
    args = args.copy()
    args['Name'] = args['Name'].str.extract('([A-Za-z]+)\.', expand=False)
    args['Name'] = args['Name'].map(name_map).fillna(0).astype(int)
    args['Sex'] = args['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
    args['Embarked'] = args['Embarked'].map({'S':0, 'C':1, 'Q':2}).astype(int)
    args['Cabin'] = args['Cabin'].map(cabin_maps).fillna(0).astype(int)
    args['Age'] = args['Age'].fillna(36)
    args['Fare'] = args['Fare'].fillna(0)
    adjust_fare(args)
    adjust_age(args)                               
    return args
#2) Call your function here
new_titanic2 = feature_extraction(new_titanic)
new_titanic2.head()
```

### Question 1
Which state has the most counties in it? (hint: consider the sumlevel key carefully! You'll need this for future questions too...)

*This function should return a single string value.*

```python
def maxcounty():
    # sumlev 50 is for counties, sumlev 40 is for states
    df = census_df[census_df['SUMLEV'] == 50]
    df = df.groupby( [ "SUMLEV", "STNAME"] ).size().to_frame(name = 'count').reset_index()
    return df.loc[df['count'].idxmax()]['STNAME']
maxcounty()
```

### Question 2
Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Use `CENSUS2010POP`.

*This function should return a list of string values.*

```python
def top_popcounties():
    df=census_df[census_df['SUMLEV'] == 50]
    df1=df.sort_values(['STNAME','CENSUS2010POP'],ascending=False).groupby('STNAME').head(3).copy()
    df2 = df1.reset_index().groupby("STNAME").sum().sort_values(['CENSUS2010POP'],ascending=False).head(3).copy()
    return list(df2.index.values)
top_popcounties()
```

### Question 3
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
def populationdelta():
    df=census_df[census_df['SUMLEV'] == 50]
    df['STDEV'] = df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012',
                      'POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].std(axis=1)
    return df.loc[df['STDEV'].idxmax()]['CTYNAME']
populationdelta()
```

### Question 4
In this datafile, the United States is broken up into four regions using the "REGION" column. 

Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

*This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).*

```python
def query():
    df=census_df[census_df['SUMLEV'] == 50]
    region1 = df['REGION'] == 1
    region2 = df['REGION'] == 2
    county_name = df['CTYNAME'].str.startswith('Washington')
    pop_estim = df['POPESTIMATE2015'] > df['POPESTIMATE2014']
    df = df[(region2|region1)&county_name&pop_estim]
    return df[['STNAME', 'CTYNAME']]
query()
```
